```python
#Actual Velocity of Return                  #Designed by Kyle Neary for AERO DESIGN PROJECT 2 (ENG2002)

#at h1 Va=0
h = 305.5    #Max Altitude of Rocket (km)
q= 1000       #Max Time
dt=0.001      #Time Interval (Affects Accuracy and time of calculation)
q=int(q/dt)  #Accounts for Time Interval
Ve = 3399    #Exhaust Velocity

Va = np.zeros(q)           #Established Initial Arrays
s = np.zeros(q)
time = np.arange(0,q)

s[0]=0
Va[0]=0

x=int(round(h,len(str(h))))    #Rounds Altitude to nearest 1km as that is the intervals in the atmospheric data


for t in range(1,q):        #Calculates Freefalling Velocity
    if x<(85):
        #print('ADD DRAG')
        Va[t] = 9.81*dt - ((0.5*A[x][3]*Va[t-1]*Va[t-1]*Cd*S1)/M)*dt + Va[t-1]
        s[t] = Va[t]*dt+s[t-1]
        x=int(h*1000-int(s[t]))
        x=int(round(x,0)/1000)
        if (h*1000-int(s[t]))<-100:
            break
        if Va[t]>V[85-x]:
            #print('At terminal Velocity')
            #print(x)
            Va[t]=V[85-x]
    else:
        Va[t] = 9.81*dt + Va[t-1]
        s[t] = Va[t]*dt+s[t-1]
        x=int(h*1000-int(s[t]))
        x=int(round(x,0)/1000)
    #print('Altitude =',h*1000-s[t])
    #print('Fall Time (s) =', round(t*dt,2),'Altitude (m) =',h*1000-s[t],Va[t])

def find_nearest(array,value):           #Finds the value in the array closed to entered value
    idx = np.searchsorted(array, value, side="left")
    if idx > 0 and (idx == len(array) or math.fabs(value - array[idx-1]) < math.fabs(value - array[idx])):
        return array[idx-1]
    else:
        return array[idx]

Total_Time_Freefall = (np.abs(s - h*1000)).argmin()          #Code Here finds time when altitude = 0
Velocity_to_Burn = Va[Total_Time_Freefall]                   #Finds velocity of rocket at 0 altitude

#Rocket Assisted Landing

Mfinal = 20000           #Ideal Final Mass of Rocket
M0 = Mfinal + 3*650*10   #Initial Mass (inc fuel)
Ve = 3399                #Nozzle Velocity
n = 0.7                  #Throttle
Thrust = 6600*1000*n     #Thrust of Rocket at decoupling

dV = int(Ve*np.log(M0/Mfinal))    #DeltaV of Returning Booster
acc = Thrust/M0                   #Acceleration of booster after decoupling

#For this report we assume a constant acceleration of the rocket as it comes into land
#This acceleration is conservativly calculated at Stage Seperation however this acc will be greater in reality
#Assume Drag is neglegable when the burn starts

accRR = acc-9.81    #Effective constant acceleration of rocket

Calc_Time = int(10/dt)   #Amount of time before total time freefall the following calculations are done for (10 seconds in this example)

BT = np.zeros(400)
BT = Va[slice(Total_Time_Freefall-Calc_Time,Total_Time_Freefall)]/accRR        #Calculates Burn time at different times of Engine ON
Distance = 0.5*BT*Va[slice(Total_Time_Freefall-Calc_Time,Total_Time_Freefall)] #Calculates the distance to ground at each BT

Test = Distance-(h*1000-s[slice(Total_Time_Freefall-Calc_Time,Total_Time_Freefall)])  #Creates an array of Distance traveled-altitude
Test_Result = int(np.where(Test == find_nearest(Test,0))[0])               #Finds when distance traveled - altitude = 0
Time_Burn_On = Total_Time_Freefall-Calc_Time+Test_Result+1                   #Calculates time of Engine On

for t in range(Time_Burn_On,q):       #Corrects Velocities during final burn
    Va[t] = Va[t-1]-accRR*dt          #Assume Constant Acceleration & no drag
    s[t] = Va[t]*dt+s[t-1]
    x=int(h*1000-int(s[t]))
    x=int(round(x,1-len(str(x)))/1000)
    #print('Burn ON','Fall Time (s)', round(t*dt,2),'Altitude',h*1000-s[t],Va[t])

Engine_Shutoff = int(np.where(Va == find_nearest(Va[slice(Time_Burn_On,q)][::-1],0))[0])     #[::-1] is used to reverse the array from descending to ascending as this is needed for the function
Landing_V = np.sqrt(Va[Engine_Shutoff]*Va[Engine_Shutoff]+2*9.81*(h*1000-s[Engine_Shutoff]))
r = np.where(Va==np.max(Va))

print('Engine Time On =',Time_Burn_On*dt,'s')
print('Engine Time On Altitude =',h*1000-s[Time_Burn_On],'m')
print('Engine Shut off Time =',Engine_Shutoff*dt,'s')
print('Engine Burn Time =',(Engine_Shutoff-Time_Burn_On)*dt,'s')
print('Total Delta V of Returning Rocket =',dV,'m/s')
print('Acceleration of Returning rocket =',int(acc),'m/s^2')
print('Maximum Descent Velocity =',int(max(Va)),'m/s')
print('Altitude of Max Velocity =',round(h*1000-int(s[r])),'m')
print('Engine Cutoff Altitude =',h*1000-s[Engine_Shutoff],'m')
print('Engine Cutoff Velocity =',Va[Engine_Shutoff],'m/s')
print('Landing Velocity of Rocket =',Landing_V,'m/s')

plt.figure(figsize=[15,10])
plt.subplot(211)
plt.plot(time[slice(0,Engine_Shutoff)]*dt,Va[slice(0,Engine_Shutoff)])
plt.title('Velocity of Returning rocket')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.grid()
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

plt.subplot(212)
plt.plot(time[slice(0,Engine_Shutoff)]*dt,(h*1000-s[slice(0,Engine_Shutoff)])/1000)
plt.title('Return Altitude Against Time')
plt.xlabel('Time (s)')
plt.ylabel('Altitude (km)')
plt.grid()
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
```
